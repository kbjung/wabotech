# Library

```python
import pandas as pd
import numpy as np
import os
```

# [로드] [ODS]자동차 등록정보(CEG_CAR_MIG)

## columns

```python
file_name = 'CEG_CAR_MIG'
col_exc1 = pd.read_excel(f'analysis/ODSvsSTD/[정제]{file_name}.xlsx', sheet_name='ODS_col')
```

```python
col_exc1
```

```python
col_exc1 = col_exc1.loc[:24]
col_exc1
```

```python
col_dict1 = {}
for a, b in zip(col_exc1['내용'].to_list(), col_exc1['컬럼'].to_list()):
    col_dict1[a] = b
col_dict1
```

```python
# 필요한 컬럼만 추출
sel_col1 = [
    '차량관리번호',
    '차량등록번호',
    '본거지 법정동 코드',
    '소유자 구분',
    '차종',
    '용도',
    '차대번호',
    '최초 등록 일자',
    '차량 연식',
    '차량 제작 일자',
    '배출가스검사유효일자',
    '제원관리번호',
    '배출가스 인증번호',
    '배출가스 등급',
    '저공해조치 종류',
    '저공해조치 구조변경 YN',
    '차량 말소 YN',
    '취득 일자',
    '취득 금액',
    ]
col1 = []
for one in sel_col1:
    col1.append(col_dict1[one])
col1
```

## [경로]

```python
ods_fold = 'D:\과업\[한국환경공단]자동차_환경_빅데이터_구축_2차\data\ODS'
ods_path = os.path.join(ods_fold, f'ODS_{file_name}.csv')
```

```python
# about 5-6m
df1 = pd.read_csv(ods_path, low_memory=False, usecols=col1)
```

```python
df1.info()
```

```python
df1.head()
```

```python
df1.columns = sel_col1
df1.head()
```

## 코드화 정보 해석

```python
df1['소유자 구분'].unique()
```

```python
# 소유자 구분
own_dict = {'B':'사업자', 'F':'외국인', 'L':'법인', 'O':'기타', 'R':'주민'}
df1['소유자 구분'] = df1['소유자 구분'].map(own_dict)
df1.head()
```

```python
df1['소유자 구분'].unique()
```

```python
df1['차종'].unique()
```

```python
# 차종
car_cate_dict = {'M':'이륜', 'P':'승용', 'S':'특수', 'T':'화물', 'V':'승합'}
df1['차종'] = df1['차종'].map(car_cate_dict)
df1['차종'].unique()
```

```python
df1['용도'].unique()
```

```python
# 용도
use_dict = {'B':'영업용', 'O':'관용', 'P':'자가용', 'T':'개인택시'}
df1['용도'] = df1['용도'].map(use_dict)
df1['용도'].unique()
```

```python
df1.head()
```

## [출력] [ODS]자동차_등록정보(CEG_CAR_MIG)

```python
# about 4m
df1.to_csv(os.path.join(ods_fold, '[ODS]자동차_등록정보(2022-12-29).csv'), index=False)
```

## [출력] [ODS][4등급]자동차_등록정보(CEG_CAR_MIG)

```python
# about 20s
g4.to_csv(os.path.join(ods_fold, '[ODS][4등급]자동차_등록정보(2022-12-29).csv'), index=False)
```
