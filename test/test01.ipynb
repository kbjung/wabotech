```python
ax = np.random.randint(1, 1000, 100)
ay = np.random.randint(1, 1000, 100)
by = np.random.randint(1, 1000, 100)
cy = np.random.randint(1, 1000, 100)
plt.bar(ax, ay, label='a', width=5, alpha=0.7)
plt.bar(ax, by, label='b', width=5, alpha=0.7)
plt.bar(ax, cy, label='c', width=5, alpha=0.7)
plt.legend(frameon=True, shadow=False, framealpha=0.3, fancybox=True);
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title('Sin(X)')
plt.fill([5, 5, 7, 7], [-0.5, 0.25, 0.25, -0.5], color='gray', alpha=0.5)
plt.plot(x, y1)

plt.subplot(1, 3, 2)
plt.fill_between(x[2: 20], y2[2: 20], alpha=0.5)
plt.title('Cos(X)')
plt.plot(x, y2)

plt.subplot(1, 3, 3)
plt.title('Tan(X)')
plt.plot(x, y3)
plt.tight_layout();
```
