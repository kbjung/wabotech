# Library

```python
import pandas as pd
import numpy as np
import os
```

# [로드] 제원정보

```python
ods_fold = 'D:\과업\[한국환경공단]자동차_환경_빅데이터_구축_2차\data\ODS'
file_name = '[ODS]자동차_제원정보(한글컬럼)(2022.12.29).csv'
ods_path = os.path.join(ods_fold, file_name)
```

```python
# about 
df = pd.read_csv(ods_path, encoding='cp949', low_memory=False)
df.info()
```

```python
col = pd.read_excel('analysis/ODSvsSTD/[정제]CEG_CAR_SRC_MIG.xlsx', sheet_name='ODS_col')
col
```

```python
col_dict = {b:a for a, b in zip(col['컬럼'], col['내용'])}
col_dict
```

# 결측치

```python
an = df.isnull().sum().reset_index()
an.columns = ['한글컬럼', '결측치']
```

```python
temp = []
for x in an['한글컬럼']:
    try:
        temp.append(col_dict[x])
    except:
        temp.append('미정')
an['컬럼'] = temp
an = an[['컬럼', '한글컬럼', '결측치']]
an
```

## [출력] 정기검사 결측치

```python
an.to_excel(f'analysis2/[BD1][통계]제원정보.xlsx', index=False)
```
