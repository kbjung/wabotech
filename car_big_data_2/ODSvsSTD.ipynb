```python
def fuc(num):
   # 파일명, 파일경로 설정
   ods_file_name = f'ODS_CEG_CAR_MIG_{num}'
   std_file_name = f'STD_CEG_CAR_MIG_{num}'
   opath = os.path.join(ods_path, ods_file_name + '.csv')
   spath = os.path.join(std_path, std_file_name + '.csv')
   odf = pd.read_csv(opath)
   sdf = pd.read_csv(spath)

   # 분석
   olist = [
   'ACQS_AMT', 
   'ACQS_DE', 
   'BSPL_LEDO_CD', 
   'CRTN_DT', 
   'EMIS_CRTCNO', 
   'EMIS_GRD', 
   'EMIS_INSP_EFCT_DE', 
   'FST_MBERNO', 
   'FST_REG_DE', 
   'LEM_KND', 
   'LEM_STRCHG_YN', 
   'LEM_YN', 
   'NOW_MBERNO', 
   'OWNR_SE', 
   'PURPS', 
   'SRCMNNO', 
   'UPDT_DT', 
   'VHCL_ERSR_YN', 
   'VHCL_MNFCT_DE', 
   'VHCL_YRIDNW', 
   'VHCTY', 
   'VHMNO', 
   'VHRNO_NUM_INFO', 
   'VIN', 'VHRNO', 
   'NOW_OWNR_NM', 
   'BSPL_ADRS', 
   'OWNR_ADRS', 
   'BSPL_ZIP', 
   'FST_OWNR_NM', 
   'RMK', 
   'USER_ID'
   ]
   slist = [
   'ACQS_AMT', 
   'ACQS_YMD', 
   'BSPL_STDG_CD', 
   'VHCL_REG_DT', 
   'EXHST_GAS_CERT_NO', 
   'EXHST_GAS_GRD_CD', 
   'EXTGAS_INSP_VLD_YMD', 
   'FRST_MBR_NO', 
   'FRST_REG_YMD', 
   'LEM_KND_CD', 
   'STRCT_CHG_YN', 
   'LEM_YN', 
   'NOW_MBR_NO', 
   'VHCL_OWNR_CL_CD', 
   'PURPS_CD2', 
   'MANP_MNG_NO', 
   'MDFCN_DT', 
   'VHCL_ERSR_YN', 
   'VHCL_FBCTN_YMD', 
   'YRIDNW', 
   'VHCTY_CD', 
   'VHCL_MNG_NO', 
   'VHRNO_FGRE_INFO', 
   'VIN', 
   'VHRNO'
   ]
   odict = {
   '컬럼':olist,
   '내용':['취득금액', '취득일자', '본거지 법정동 코드', '생성 일시', '배출가스 인증번호', '배출가스 등급', '배출가스검사유효일자', '최초 회원번호', '최초 등록 일자', '저공해조치 종류', '저공해조치 구조변경 YN', '저공해조치 YN(미사용)', '현재 회원번호', '소유자 구분', '용도', '제원관리번호', '수정 일시', '차량 말소 YN', '차량 제작 일자', '차량 연식', '차종', '차량관리번호', '차량등록번호 숫자 정보', '차대번호', '미정', '미정', '미정', '미정', '미정', '미정', '미정', '미정']
   }
   ocol = pd.DataFrame(odict)

   sdict = {
      '컬럼':slist,
      '내용':['취득금액', '취득일자', '본거지법정동코드', '차량등록일시', '배출가스인증번호', '배출가스등급', '배출가스검사유효일자', '최초회원번호', '최초등록일자', '저공해조치종류코드', '구조변경여부', '저공해조치여부', '현재회원번호', '차량소유자분류코드', '용도코드2', '제원관리번호', '수정일시', '차량말소여부', '차량제작일자', '연식', '차종코드', '차량관리번호', np.nan, '차대번호', '미정']
      }
   scol = pd.DataFrame(sdict)

   odf2 = odf[olist]
   sdf2 = sdf[slist]

   ocount = odf2.count().reset_index()
   ocount.columns = ['index', 'ocount']

   scount = sdf2.count().reset_index()
   scount.columns = ['index', 'scount']
   oscount = pd.concat([ocount, scount], axis=1)
   oscount['차이'] = oscount['ocount'] - oscount['scount']

   # 분석 결과 파일로 출력
   texcel = pd.ExcelWriter(f'analysis/ODSvsSTD/[정제]{ods_file_name}_vs_{std_file_name}.xlsx')
   ocol.to_excel(texcel, sheet_name='ODS_col', index=False)
   scol.to_excel(texcel, sheet_name='STD_col', index=False)
   oscount.to_excel(texcel, sheet_name='count', index=False)
   # odf2.count().reset_index().to_excel(texcel, sheet_name='ODS_count',index=False)
   # sdf2.count().reset_index().to_excel(texcel, sheet_name='STD_count',index=False)
   texcel.save()

   return print(f'[정제]{ods_file_name}_vs_{std_file_name}.xlsx 출력 완료.')
```
