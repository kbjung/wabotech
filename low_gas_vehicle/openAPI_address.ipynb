### 지역코드 수집 함수화

```python
def func_zcode(max_page, step, zcode):
    df = pd.DataFrame()
    step = step - 1
    for n in tqdm(range(max_page - step, max_page + 1)):
        serviceKey = 'dPa1i52mLHoVyeAvr9pW8EFAE9Zd99BNg%2FzvGLn1wm716miCQuLuMlRPXJVIpNLK%2Fyi0QoTDM2RXRlfQphK%2Fig%3D%3D'
        numOfRows = 10
        pageNo = n
        zcode = zcode
        url = f'http://apis.data.go.kr/B552584/EvCharger/getChargerInfo?serviceKey={serviceKey}&numOfRows={numOfRows}&pageNo={pageNo}&dataType=XML&zcode={zcode}'
        result = requests.get(url)
        soup = bs(result.text, 'xml')
        items = soup.find_all('item')

        a, a1, a2, z, zs = [], [], [], [], []
        for one in items:
            atemp = one.find("addr").text
            if atemp == 'null':
                a.append(np.nan)
                a1.append(np.nan)
                a2.append(np.nan)
            else:
                a.append(atemp)
                a1.append(atemp.split()[0])
                a2.append(atemp.split()[1])
            ztemp = one.find("zcode").text
            if ztemp == 'null':
                z.append(np.nan)
            else:
                z.append(ztemp)
            zstemp = one.find("zscode").text
            if zstemp == 'null':
                zs.append(np.nan)
            else:
                zs.append(zstemp.replace(ztemp, '', 1))
        
        tdf = pd.DataFrame({'address':a, 'ad1':a1, 'code1':z, 'ad2':a2, 'code2':zs})
        df = pd.concat([df, tdf], ignore_index=True)
    print(f'지역코드 : {zcode}, {max_page - step}~{max_page}페이지까지 수집한 데이터프레임')
    return df
```
