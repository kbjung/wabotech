```python
import pandas as pd
from tqdm import tqdm
```

# test

```python
a = {'A' : [1, 2, 3, 4, 5, 4], 'B' : [10, 20, 30, 40, 50, 60], 'C' : [100, 200, 300, 400, 500, 400], 'D' : [1000, 2000, 3000, 4000, 5000, 4000]}
test = pd.DataFrame(a)
test
```

```python
test.loc[(test['A'] == 4) & (test['C'] == 400) & (test['D'] == 4000), ['A', 'B', 'C']] = [100, 1, 2]
test 
```

```python
for one, two, three in tqdm(zip(test['A'], test['B'], test['C'])):
    print(one, two, three)
```

```python
b = {'A' : [1, 2, 3, 4, 5, 4], 'B2' : [10, 20, 30, 40, 50, 60], 'C' : [100, 200, 300, 400, 500, 400], 'D2' : [1000, 2000, 3000, 4000, 5000, 4000]}
test2 = pd.DataFrame(b)
test2
```

```python
ttest = pd.concat([test, test2], ignore_index=True)
ttest
```
