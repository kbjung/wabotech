# Library

```python
import pandas as pd
import numpy as np
import os
```

# [로드] 정기검사

```python
ods_fold = 'D:\과업\[한국환경공단]자동차_환경_빅데이터_구축_2차\data\ODS'
file_name = '[ODS][G4]정기검사(말소제거)(중복차대제거)(2022.12.29).csv'
ods_path = os.path.join(ods_fold, file_name)
```

```python
# about 2m
df = pd.read_csv(ods_path, low_memory=False)
df.info()
```

```python
col = pd.read_excel('analysis/ODSvsSTD/[정제]TB_JGT_HIS.xlsx', sheet_name='ODS_col')
col
```

```python
col_dict = {b:a for a, b in zip(col['컬럼'], col['내용'])}
col_dict
```

# 결측치

```python
an = df.isnull().sum().reset_index()
an.columns = ['한글컬럼', '결측치']
an['컬럼'] = [col_dict[x] for x in an['한글컬럼']]
an = an[['컬럼', '한글컬럼', '결측치']]
an
```

## [출력] 정기검사 결측치

```python
an.to_excel(f'analysis2/[BD1][통계]정기검사(말소제거)(중복차대제거).xlsx', index=False)
```
