# seaborn setting

## 전체 스타일링
- context
    - notebook: 기본 설정
    - paper: 논문, 보고서
    - talk: 프리젠테이션
    - poster: 포스터
- style
    - darkgrid: 회색 배경색 + 그리드
    - whitegrid: 흰색 배경색 + 그리드
    - dark: 회색 배경색
    - white: 흰색 배경색
    - ticks: 흰색 배경색 + 눈금
- palette
    - deep
    - muted
    - pastel
    - bright
    - dark
    - colorblind

```python
for one in ['deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind']:
    sns.palplot(sns.color_palette(one))
```

```python
sequential_colors = sns.color_palette("RdPu", 10)
sns.palplot(sequential_colors)
```

```python
diverging_colors = sns.color_palette("RdBu", 10)
sns.palplot(diverging_colors)
```

```python
sns.set_theme(
    context='notebook',   # 매체: paper, talk, poster
    style='darkgrid',     # 기본 내장 테마
    palette='deep',       # 그래프 색
    font='D2Coding', # 글꼴 종류 
    # font='Malgun Gothic', # 글꼴 종류 
    font_scale=1,         # 글꼴 크기
    # rc=custom_params # 그래프 세부 사항
    )     
```

#### subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20,10))
plt.suptitle('flights')
sns.barplot(data=fg, x='year', y='passengers', ax=axes[0, 0])
axes[0, 0].set_title('mean')
sns.barplot(data=fg, x='year', y='passengers', estimator=np.median, ax=axes[0, 1])
axes[0, 1].set_title('median')
sns.boxplot(fg, x='year', y='passengers', ax=axes[1, 0])
axes[1, 0].set_title('box')
sns.lineplot(data=fg, x='year', y='passengers', ax=axes[1, 1])
axes[1, 1].set_title('line');
```

#### lmplot

```python
sns.set_style('darkgrid')
sns.lmplot(x='year', y='passengers', data=fg, height=7);
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='year', y='passengers', hue='month', data=fg, height=7);
```

#### catplot

```python
sns.catplot(df, x='class', y='fare');
```

```python
sns.catplot(df, x='class', kind='count');
```

```python
sns.catplot(df, x='class', y='fare', kind='bar');
```

```python
sns.catplot(df, x='class', y='fare', kind='bar', estimator=np.median);
```

```python
sns.catplot(df, x='class', y='fare', kind='bar', estimator=len);
```

```python
# catplot은 subplots 안먹힘.
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
sns.catplot(df, x='class', y='fare', kind='bar')
ax1.set_title('mean')
ax2 = fig.add_subplot(1, 2, 2)
sns.catplot(df, x='class', y='fare', kind='bar', estimator=np.median)
ax2.set_title('median')
```

```python
# catplot은 subplots 안먹힘.
fig, axes = plt.subplots(ncols=2)
sns.catplot(data=df, x='class', y='fare', kind='bar', ax=axes[0])
axes[0].set_title('mean')
sns.catplot(data=df, x='class', y='fare', kind='bar', estimator=np.median, ax=axes[1])
axes[1].set_title('median')
```
